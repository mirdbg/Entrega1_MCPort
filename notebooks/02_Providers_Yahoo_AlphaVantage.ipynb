```

import sys
from pathlib import Path
base = Path.cwd(); src = base / "src"
if str(src) not in sys.path: sys.path.insert(0, str(src))
from mcport import PriceSeries
import pandas as pd
from mcport.providers import YahooProvider, AlphaVantageProvider
```

```

# Yahoo Finance
try:
    data = YahooProvider().price_history(symbols=["AAPL","MSFT"], periods=90)
    for k, df in data.items():
        print("Ticker:", k); display(df.head())
except Exception as e:
    print("Yahoo offline:", e)
```
